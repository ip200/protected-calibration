# Plots of calibrating functions

This is the code for Figure 1.
Quadratic and Cox $\alpha$ are similar to each other.
And cubic and Cox $\beta$ are similar to each other.

```python
import numpy as np
import matplotlib.pyplot as plt

# parameters:
theta = 1    # quadratic
a = 1.5      # cubic
b = 0.5      # cubic
alpha = 1    # Cox alpha
beta = 0.75  # Cox beta

def quadratic(p):
  return p+theta*p*(1-p)

def cubic(p):
  return p+a*p*(p-b)*(p-1)

def Cox_alpha(p):
  return p*np.exp(alpha)/(p*np.exp(alpha)+(1-p))

def Cox_beta(p):
  return p**beta/(p**beta+(1-p)**beta)

x = np.arange(0,1,0.01)
y_quad = np.empty(x.size)
y_cub = np.empty(x.size)
y_Cox_alpha = np.empty(x.size)
y_Cox_beta = np.empty(x.size)

for n in range(x.size):
  y_quad[n] = quadratic(x[n])
  y_cub[n] = cubic(x[n])
  y_Cox_beta[n] = Cox_beta(x[n])
  y_Cox_alpha[n] = Cox_alpha(x[n])

plt.plot(x,y_quad,'r',label='quadratic')
plt.plot(x,y_Cox_alpha,'orange',label=r'Cox $\alpha$')
plt.plot(x,y_cub,'b',label='cubic')
plt.plot(x,y_Cox_beta,'g',label=r'Cox $\beta$')
plt.plot(x,x,':')
plt.legend()

## Save the figure
# plt.savefig('cal_functions.pdf', bbox_inches='tight')
```
